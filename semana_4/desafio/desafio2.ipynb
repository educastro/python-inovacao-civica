Escolha a página do wikipédia relacionada a lista de parlamentares do seu estado e extraia as tabelas existentes na página.

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'


url = "https://pt.wikipedia.org/wiki/Lista_de_parlamentares_do_Distrito_Federal_(Brasil)"

tables = pd.read_html(url)
pd.set_option('display.max_rows', None)
```

### Quantas tabelas existem na página?

```python
for id, table in enumerate(tables):
    print(f'Table {id}:')
    print(table.info())
    print('\n')
```

### Qual o total de senadores?

```python
senadores_table = tables[0][:-1]
senadores_table
```

### Qual o total de deputados federais?

```python
deputados_table = tables[1][:-1]
deputados_table
```

### Qual o total de deputados e senadores por cidade?

dica: `groupby`

```python
senadores_table_copy = senadores_table.copy()
deputados_table_copy = deputados_table.copy()

senadores_table_copy["Cargo"] = "Senador"
senadores_table_copy.rename(columns={"Senadores eleitos": "Nome do parlamentar"}, inplace=True)
deputados_table_copy["Cargo"] = "Deputado Federal"
deputados_table_copy.rename(columns={"Deputados federais eleitos": "Nome do parlamentar"}, inplace=True)

parlamentares_table = pd.concat([senadores_table_copy, deputados_table_copy], ignore_index=True)
parlamentares_table.groupby(["Naturalidade", "Cargo"])["Nome do parlamentar"].count().sort_values(ascending=False).to_frame()
```

### Qual deputado/senador com mais mandatos?

```python
parlamentares_table.groupby(["Nome do parlamentar", "Cargo"])["Mandatos"].sum().sort_values(ascending=False)
```
